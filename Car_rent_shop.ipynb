```python
cars = {'Car1': {'id': 1, 'name': 'Hyundai', 'charges_per_hour': 500, 'available': True},
        'Car2': {'id': 2, 'name': 'Maruti', 'charges_per_hour': 350, 'available': True},
        'Car3': {'id': 3, 'name': 'Ford', 'charges_per_hour': 700, 'available': True},
        'Car4': {'id': 4, 'name': 'Kia', 'charges_per_hour': 600, 'available': True},
        'Car5': {'id': 5, 'name': 'BMW', 'charges_per_hour': 800, 'available': True}}

rent_records = []
earning = 0

while True:
    print("Welcome to Car Rental Shop!")
    print("1. Rent a Car")
    print("2. Display Available Cars")
    print("3. Return a Car")
    print("4. Rented car Details")
    print("5. Earnings made")
    print("6. Exit")

    option = input("Enter option no: ")

    # Car rent
    if option == "1":
        print("Available Cars :")
        for name, info in cars.items():
            if info['available']:
                print(info['id'], info['name'], "(Charge per hour: Rs", info['charges_per_hour'])

        rent_car = input("Enter the car no (Car1,Car2..) you want to rent: ")
        if rent_car in cars:
            customer_name = input("Enter your name: ")
            time = int(input("Enter the number of hours you want to rent the car: "))
            bill = cars[rent_car]["charges_per_hour"] * time

            rent_records.append({
                "customer_name": customer_name,
                "car": rent_car,
                "time": time,
                "bill": bill
            })

            cars[rent_car]["available"] = False
            earning = earning + bill
            print(f"{customer_name} has rented {rent_car} for {time} hours. Your bill is Rs {bill}.")
        else:
            print("Invalid choice or the car is not available. Please try again.")

    # Available cars
    elif option == "2":
        available = []
        for car, info in cars.items():
            if info["available"]:
                available.append(car)
        if available:
            print("Available Cars:")
            for car in available:
                print(f"{car}")
        else:
            print("Sorry, no cars are available for rent at the moment.")

    # Return car
    elif option == "3":
        return_id = input("Enter Car ID (Car1, Car2, ...) you want to return: ")
        if return_id in cars:
            car_info = cars[return_id]
            if not car_info['available']:
                for record in rent_records:
                    if record['car'] == return_id:
                        rent_records.remove(record)
                        car_info['available'] = True
                        print("Car returned Successfully.")
                        break
                else:
                    print("Car not returned. No rental record found.")
            else:
                print("Car is already available.")
        else:
            print("Car not found. Please enter a valid car ID.")

    # Rent Records
    elif option == "4":
        if rent_records:
            print("Rental Records:")
            for record in rent_records:
                print(f"Car Details: {record['car']}")
                print(
                    f" Customer Details: (Rented by {record['customer_name']}, Hours: {record['time']}, Bill: Rs {record['bill']})")
        else:
            print("No cars have been rented yet.")

    # earnings
    elif option == "5":

        print(f"Total earnings: Rs {earning}")

    elif option == "6":
        # Exit the program
        print("Thank you for using the Car Rental Shop. Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")

# End of the program
```
